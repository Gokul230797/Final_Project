```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, learning_curve
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, ConfusionMatrixDisplay, accuracy_score
from xgboost import XGBClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical


# -------------------------------
# 1. Data Preprocessing
# -------------------------------
def preprocess_data(file_path):
    data = pd.read_csv(file_path)

    # Check missing values, duplicates, and data info
    print("Dataset Overview:\n", data.info())
    print("\nMissing Values:\n", data.isnull().sum())
    print(f"\nDuplicate Rows: {data.duplicated().sum()}")
    data = data.drop_duplicates()

    # Encode RiskLevel
    label_encoder = LabelEncoder()
    data['RiskLevel'] = label_encoder.fit_transform(data['RiskLevel'])
    print(f"\nRisk Level Encoding: {dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))}")

    # Feature Engineering
    data['PulsePressure'] = data['SystolicBP'] - data['DiastolicBP']
    data['RiskAgeRatio'] = data['Age'] / (data['HeartRate'] + 1)
    return data, label_encoder


# -------------------------------
# 2. Exploratory Data Analysis (EDA)
# -------------------------------
def perform_eda(data):
    """
    Performs Exploratory Data Analysis (EDA) with multiple visualizations.
    """
    print("\nPerforming Exploratory Data Analysis...")

  # 1. Age Distribution with Risk Level Overlay
    plt.figure(figsize=(8, 6))
    sns.histplot(data=data, x='Age', hue='RiskLevel', multiple='stack', palette='Set2', kde=True)
    plt.title('Age Distribution by Risk Level')
    plt.xlabel('Age (Years)')
    plt.ylabel('Count')
    plt.savefig('Age Distribution by Risk Level.png', dpi=300)
    plt.show()

    # 2. Blood Sugar Levels by Risk Level (Violin Plot)
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=data, x='RiskLevel', y='BS', palette='coolwarm', inner='quartile')
    plt.title('Blood Sugar Distribution by Risk Level')
    plt.xlabel('Risk Level')
    plt.ylabel('Blood Sugar (mmol/L)')
    plt.savefig('Blood Sugar Distribution by Risk Level.png', dpi=300)
    plt.show()

    # 3. Systolic vs Diastolic Blood Pressure (Scatter Plot)
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x='SystolicBP', y='DiastolicBP', hue='RiskLevel', palette='Set1', s=100)
    plt.title('Systolic vs. Diastolic Blood Pressure by Risk Level')
    plt.xlabel('Systolic BP (mm Hg)')
    plt.ylabel('Diastolic BP (mm Hg)')
    plt.savefig('Systolic vs. Diastolic Blood Pressure by Risk Level.png', dpi=300)
    plt.show()

    # 4. Pairplot for Key Features (Pairplot)
    plt.figure(figsize=(8, 6))
    selected_features = ['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'HeartRate', 'RiskLevel']
    sns.pairplot(data[selected_features], hue='RiskLevel', palette='husl', diag_kind='kde')
    plt.suptitle('Pairplot of Key Features', y=1.02)
    plt.savefig('Pairplot of Key Features.png', dpi=300)
    plt.show()

    # 5. Correlation Heatmap
    correlation_matrix = data.corr()
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
    plt.title('Correlation Heatmap')
    plt.savefig('Correlation Heatmap.png', dpi=300)
    plt.show()

    # 6. Distribution of Heart Rate by Risk Level (Boxplot)
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, x='RiskLevel', y='HeartRate', palette='coolwarm')
    plt.title('Heart Rate Distribution by Risk Level')
    plt.xlabel('Risk Level')
    plt.ylabel('Heart Rate (bpm)')
    plt.savefig('Heart Rate Distribution by Risk Level.png', dpi=300)
    plt.show()


# -------------------------------
# 3. Train-Test Split
# -------------------------------
def prepare_data(data):
    X = data[['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate', 'PulsePressure', 'RiskAgeRatio']]
    y = data['RiskLevel']

    # Handle class imbalance
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)

    # Feature scaling
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_resampled)

    # Split data
    return train_test_split(X_scaled, y_resampled, test_size=0.2, random_state=42)



# -------------------------------
# 4. Random Forest Model
# -------------------------------
def random_forest_model(X_train, X_test, y_train, y_test):
    param_grid = {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }
    grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    y_pred = best_model.predict(X_test)
    print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred))

    ConfusionMatrixDisplay.from_estimator(best_model, X_test, y_test, cmap='Blues', values_format='d')
    plt.title("Random Forest Confusion Matrix")
    plt.savefig('Random Forest Confusion Matrix.png', dpi=300)
    plt.show()


# -------------------------------
# 5. XGBoost Model
# -------------------------------
def xgboost_model(X_train, X_test, y_train, y_test):
    param_grid = {
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 7],
        'n_estimators': [50, 100, 200],
        'subsample': [0.8, 1]
    }
    grid_search = GridSearchCV(XGBClassifier(eval_metric='logloss', random_state=42), param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    y_pred = best_model.predict(X_test)
    print("\nXGBoost Classification Report:\n", classification_report(y_test, y_pred))

    ConfusionMatrixDisplay.from_estimator(best_model, X_test, y_test, cmap='Blues', values_format='d')
    plt.title("XGBoost Confusion Matrix")
    plt.savefig('XGBoost Confusion Matrixl.png', dpi=300)
    plt.show()


# -------------------------------
# 6. ANN Model
# -------------------------------
def ann_model(X_train, X_test, y_train, y_test):
    y_train_categorical = to_categorical(y_train)
    y_test_categorical = to_categorical(y_test)

    model = Sequential()
    model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
    model.add(Dropout(0.3))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(y_train_categorical.shape[1], activation='softmax'))
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    model.fit(X_train, y_train_categorical, validation_split=0.2, epochs=50, batch_size=32, verbose=1)

    y_pred = np.argmax(model.predict(X_test), axis=1)
    print("\nANN Classification Report:\n", classification_report(y_test, y_pred))

    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
    plt.title("ANN Confusion Matrix")
    plt.savefig('ANN Confusion Matrix.png', dpi=300)
    plt.show()

# -------------------------------
# 7. Learning Curve for Random Forest and XGBoost
# -------------------------------
def plot_learning_curve_with_scores(estimator, X, y, cv, scoring, title):
    """
    Plots learning curve with training and validation scores, including score annotations.
    """
    train_sizes, train_scores, val_scores = learning_curve(
        estimator, X, y, cv=cv, scoring=scoring, train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
    )

    # Compute mean and std deviation
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    val_mean = np.mean(val_scores, axis=1)
    val_std = np.std(val_scores, axis=1)

    # Plot learning curve
    plt.figure(figsize=(10, 6))
    plt.plot(train_sizes, train_mean, label="Training Score", color="blue", marker="o")
    plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="blue", alpha=0.2)
    plt.plot(train_sizes, val_mean, label="Validation Score", color="orange", marker="o")
    plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, color="orange", alpha=0.2)

    # Annotate scores
    for i, size in enumerate(train_sizes):
        plt.text(size, train_mean[i], f"{train_mean[i]:.2f}", color="blue", ha="center")
        plt.text(size, val_mean[i], f"{val_mean[i]:.2f}", color="orange", ha="center")

    plt.title(title)
    plt.xlabel("Training Set Size")
    plt.ylabel("Accuracy")
    plt.legend(loc="lower right")
    plt.grid()
    plt.savefig(f"{title.replace(' ', '_')}.png", dpi=300)
    plt.show()


# -------------------------------
# 8. ANN Learning Curve
# -------------------------------
def plot_ann_learning_curve_with_scores(X_train, y_train, X_test, y_test, train_sizes, runs=3):
    """
    Plots a learning curve for ANN with training and validation scores.
    """
    train_scores = np.zeros((len(train_sizes), runs))
    val_scores = np.zeros((len(train_sizes), runs))

    for run in range(runs):
        for i, size in enumerate(train_sizes):
            # Subset data
            subset_size = int(size * len(X_train))
            X_train_subset = X_train[:subset_size]
            y_train_subset = y_train[:subset_size]

            # Define ANN model
            model = Sequential()
            model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
            model.add(Dropout(0.3))
            model.add(Dense(64, activation='relu'))
            model.add(Dropout(0.3))
            model.add(Dense(32, activation='relu'))
            model.add(Dense(y_train.shape[1], activation='softmax'))
            model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

            # Train model
            model.fit(X_train_subset, y_train_subset, epochs=20, verbose=0, batch_size=32)

            # Evaluate
            train_pred = model.predict(X_train_subset)
            val_pred = model.predict(X_test)
            train_scores[i, run] = accuracy_score(np.argmax(y_train_subset, axis=1), np.argmax(train_pred, axis=1))
            val_scores[i, run] = accuracy_score(np.argmax(y_test, axis=1), np.argmax(val_pred, axis=1))

    # Plot learning curve
    plt.figure(figsize=(10, 6))
    train_mean = train_scores.mean(axis=1)
    val_mean = val_scores.mean(axis=1)
    plt.plot(train_sizes, train_mean, label="Training Score", marker="o", color="blue")
    plt.plot(train_sizes, val_mean, label="Validation Score", marker="o", color="orange")

    for i, size in enumerate(train_sizes):
        plt.text(size, train_mean[i], f"{train_mean[i]:.2f}", color="blue", ha="center")
        plt.text(size, val_mean[i], f"{val_mean[i]:.2f}", color="orange", ha="center")

    plt.title("Learning Curve: ANN")
    plt.xlabel("Training Set Size")
    plt.ylabel("Accuracy")
    plt.legend(loc="lower right")
    plt.grid()
    plt.savefig('Learning Curve ANN.png', dpi=300)
    plt.show()

# -------------------------------
# 9. Main Execution
# -------------------------------
if __name__ == "__main__":
    # Preprocessing
    data, label_encoder = preprocess_data('/Users/user1/Documents/Final Project/Maternal Health Risk Dataset.csv')

    # EDA
    perform_eda(data)

    # Prepare data
    X_train, X_test, y_train, y_test = prepare_data(data)

    # Model Training
    random_forest_model(X_train, X_test, y_train, y_test)
    xgboost_model(X_train, X_test, y_train, y_test)
    ann_model(X_train, X_test, y_train, y_test)
    
    # Random Forest Learning Curve
    rf_model = RandomForestClassifier(random_state=42)
    plot_learning_curve_with_scores(rf_model, X_train, y_train, cv=5, scoring="accuracy", title="Learning Curve: Random Forest")

    # XGBoost Learning Curve
    xgb_model = XGBClassifier(eval_metric='logloss', random_state=42)
    plot_learning_curve_with_scores(xgb_model, X_train, y_train, cv=5, scoring="accuracy", title="Learning Curve: XGBoost")

    # ANN Learning Curve
    y_train_cat = to_categorical(y_train)
    y_test_cat = to_categorical(y_test)

    train_sizes = np.linspace(0.1, 1.0, 10)
    plot_ann_learning_curve_with_scores(X_train, y_train_cat, X_test, y_test_cat, train_sizes, runs=3)
```
